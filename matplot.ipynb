```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([1,3,6,9])
ypoints = np.array([4,6,8,9])
plt.plot(xpoints,ypoints)
plt.show()                
```

```python
plt.plot(ypoints)
```

```python
plt.plot(xpoints,ypoints,marker = "3")
```

```python
plt.plot(ypoints,"o--r")
```

```python
plt.plot(ypoints, ls = ":",marker ="o",ms = 10, mec = "r", mfc="k")
```

```python
plt.plot(ypoints, ls ="--", c ="b",lw = "10")
plt.plot(xpoints, ls ="--", c ="r",lw = "10")
font1 ={"family":"serif","color":'darkred','size':15}
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("the plot of axis",fontdict=font1,loc = "right")
plt.show()
```

```python
plt.grid(axis ="y",ls = "--", color = "r", lw = 0.5)
plt.plot(xpoints,ypoints)
```

```python
plt.subplot(2,2,1)
plt.plot(xpoints)
plt.subplot(2,2,2)
plt.plot(ypoints)
plt.subplot(2,2,3)
plt.show()
```

```python
colors = np.array([10,20,30,40])
plt.scatter(xpoints,ypoints, c = colors,cmap = "viridis", s= [10, 10,20.5,3], alpha = 0.5)
plt.scatter(xpoints,xpoints)
```

```python
plt.bar(xpoints,ypoints,color=['red','blue','black','pink'],width = 1.5)
plt.barh(xpoints,ypoints,color='red',height = 1.5)
```

```python
x= np.random.normal(150,20,150)
plt.hist(x)
plt.show()
```

```python
y =np.array([12,34,56,19])
mylables = ['a','b','c','d']
myexplode = [0.2,0,0,0.1]
plt.pie(y,labels =mylables,explode = myexplode,shadow =True)
plt.legend(title = "the pic of pie")
plt.show()
```
